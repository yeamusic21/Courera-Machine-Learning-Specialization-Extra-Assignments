# Simple Linear Regression using Gradient Descent and Least Squares

## The point of this workbook is to create a line and add fake noise, then solve for the true line using gradient descent.  

## After this, we'll compute gradient descent in batches so we can watch as the algorithm solves for the true line.


## Part 1:

## Create Some Fake Data to Fit

```python
#### CREATE FAKE DATA

# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# simulate some data
column1 = np.repeat(1, 100)
column2 = np.arange(10,110,1)
X = np.matrix([column1,column2])
X = X.transpose()
# make up some parameters
B = np.matrix(np.array([9.345,3.961]))
B = B.transpose()
# create some fake noise
e = np.matrix(np.random.normal(0, 10, 100))
e = e.transpose()
# finally, create fake response data
Y = X*B + e

# plot our fake data to make sure it looks as we would expect (linear relationship with noise)
plt.scatter(X[:,1], Y)
plt.title('Fake Data Scatter Plot')
plt.xlabel("Data (X)")
plt.ylabel("Response (Y)")
plt.show()
```

## Solve For Coefficients Using Gradient Descent & Least Squares

```python
#### SOLVE FOR COEFFICIENTS USING GRADIENT DESCENT & LEAST SQUARES

# initialize coefficients 
Bhat = np.matrix(np.array([1,1]))
Bhat = Bhat.transpose()
# initialize convergence flag
convergence_flag = 0
# initalize iteration counter
iterations = 0

while convergence_flag == 0:
    iterations = iterations + 1
    # compute RSS   
    residuals = Y-(X*Bhat)
    RS = np.square(residuals)
    RSS = RS.sum()
    # update coefficients
    residuals = Y-(X*Bhat)
    residual_sum = residuals.sum()
    Bhat_new_0 = Bhat[0,:] + 2*(0.0000001)*(residual_sum)
    residual_sum = residuals.transpose()*X[:,1]
    Bhat_new_1 = Bhat[1,:] + 2*(0.0000001)*(residual_sum)
    Bhat_new = np.matrix([Bhat_new_0[0,0],Bhat_new_1[0,0]])
    Bhat_new = Bhat_new.transpose()
    # check for convergence
    # compute RSS   
    residuals = Y-(X*Bhat_new)
    RS = np.square(residuals)
    RSS_new = RS.sum()
    if abs(RSS_new-RSS)<0.00001:
        convergence_flag = 1
    # replace old estimate with the new one
    Bhat = Bhat_new

# print estimated coefficients     
print "The estimated Coefficients are:"
print Bhat
# print true coefficients
print "The true Coefficients are:"
print B

plt.scatter(X[:,1], Y)
plt.plot(X[:,1], X*Bhat, linewidth=2.0)
plt.title('Search Scatter Plot')
plt.xlabel("Data (X)")
plt.ylabel("Response (Y)")
plt.show()

print "total iterations to convergence:"
iterations
```
